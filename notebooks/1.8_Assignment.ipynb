```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]})
threshold = 30
df
```

```python
filtered_df = df[df.gt(threshold).any(axis=1)]
filtered_df
```
